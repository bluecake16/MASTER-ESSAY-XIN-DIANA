```python
#Load data from csv
load_file_entsoe = pd.read_csv('entsoe_load_20200101_20251231.csv',sep=',',index_col=0,parse_dates=True)
load_file_entsoe.index = pd.to_datetime(load_file_entsoe.index, utc=True).tz_convert("Europe/Berlin")
load_file_entsoe.info()
```
